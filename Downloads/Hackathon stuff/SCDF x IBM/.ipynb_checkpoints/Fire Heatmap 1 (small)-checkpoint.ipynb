```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
plt.rcParams['animation.ffmpeg_path'] = '/usr/local/bin/ffmpeg'
#gotta manually set the path for some reason

#manimation.writers.list()
```

```python
data = pd.read_csv('data.csv')
df = data[['x', 'y','risk']]
```

```python
heatmap = pd.pivot_table(df, values='risk', 
                     index=['x'], 
                     columns='y')
```

```python
cmap = sns.diverging_palette(220, 20, sep=20, as_cmap=True)
#heat_map = sns.heatmap(data, annot=True)
sns.heatmap(heatmap, cmap=cmap)
```

```python
%matplotlib notebook 

def get_data(table,rownum,title):
    data = pd.DataFrame(table.loc[rownum][2:]).astype(float)
    data.columns = {title}
    return data

title = 'Fire Risk'

d = get_data(data,18,title)
x = np.array(d.index)
y = np.array(d['Fire Risk'])
fires = pd.DataFrame(y,x)
#XN,YN = augment(x,y,10)
#augmented = pd.DataFrame(YN,XN)
fires.columns = {title}

```

```python
Writer = animation.writers['ffmpeg']
writer = Writer(fps=20, metadata=dict(artist='Me'), bitrate=1800)
```

```python
fig = plt.figure(figsize=(10,6))
plt.xlim(0, 10)
plt.ylim(0, 10) 
plt.xlabel('Year',fontsize=20)
plt.ylabel(title,fontsize=20)
plt.title('Heroin Overdoses per Year',fontsize=20)
```
